```python
# Assign the dataframe values from the previous cell into a new dataframe variable
# and make a bar plot with the text values horizontally and the numeric values as the bar heights
hospitals_df.loc[0:10,['Rating.Overall','Facility.Name']].plot(kind='bar', x='Facility.Name', y='Rating.Overall')
```

```python
# Re-do the plot from the previous cell as a horizontal bar plot
hospitals_df.loc[0:10,['Rating.Overall','Facility.Name']].plot(kind='barh', x='Facility.Name', y='Rating.Overall')
```

```python
# Re-do the plot from the previous cell
# and change at least two aesthetic elements (colors, labels, titles, ...)
barcolors = []
for i in hospitals_df.loc[0:10,['Rating.Overall','Facility.Name']].index:
    if hospitals_df.loc[i,'Facility.Name'] in ['Dale Medical Center','Helen Keller Memorial Hospital']:
            barcolors.append('red')
    else:
            barcolors.append('blue')
print(barcolors)
```

```python
import matplotlib.pyplot as plt
```

```python
ax = hospitals_df.loc[0:10,['Rating.Overall','Facility.Name']].plot(kind='barh', 
                                                                    x='Facility.Name', 
                                                                    y='Rating.Overall',
                                                                   color=barcolors)
legend = ax.legend()
legend.remove()                                               
plt.xlabel('Overall Hospital Ratings')
```
